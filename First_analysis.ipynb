```python
airpm10.to_csv( r'\Users\ione\Desktop\pm10_1.csv', index = None, header=True) 
```

```python
airpm10['variable'] = airpm10['variable'].map(lambda x: x.lstrip('H'))
```

```python
airpm10['variable']= pd.to_numeric(airpm10['variable']).astype(int)
```

```python
airpm10 = airpm10\
    .rename({"variable": "HORA", "value": "VALOR"}, axis=1)
airpm10.tail()
```

```python
airpm10 = airpm10.sort_values(by = ['DATA','HORA'], ascending = False)
airpm10.tail()
```

```python
plt.figure(figsize=(20,10))
plt.plot(airpm10["DATA"], airpm10["VALOR"])
plt.tight_layout()
```

```python
airpm10.to_csv( r'\Users\ione\Desktop\pm10_2.csv', index = None, header=True) 
```
